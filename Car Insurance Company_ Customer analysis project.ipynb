```python
x=df.drop('Customer Lifetime Value',axis=1)
y=df['Customer Lifetime Value']
```

```python
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
```

```python
encoder=ColumnTransformer([('one',OneHotEncoder(),[0,1,2,3,4,6,7,13,14,15,16,18,19]),
                           ('scal',MinMaxScaler(),[5,8,9,10,11,12,17])],remainder='passthrough')
```

```python
newx=encoder.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(newx,y,train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(ytest,ypred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model1=RandomForestRegressor(n_estimators=6)
```

```python
model1.fit(xtrain,ytrain)
ypr=model1.predict(xtest)
```

```python
print(r2_score(ytest,ypr))
```
